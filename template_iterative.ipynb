```python
%autoreload 2
import matplotlib.pyplot as plt
import numpy as np
from src import gauss_seidel  
from src import gauss_jacobi  
A = [[2, 10], [3, 2]]
b = [16, 11]
x0 = [1, 1]
x2 = [5, -2]
tol = 1e-1
max_iterations = 100

x_jacobi, tray_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=max_iterations) #x0
x_seidel, tray_seidel = gauss_seidel(A=A, b=b, x0=x0, tol=tol, max_iter=max_iterations) #x0 

x_jacobi1, tray_jacobi1 = gauss_jacobi(A=A, b=b, x0=x2, tol=tol, max_iter=max_iterations) #x2
x_seidel1, tray_seidel1 = gauss_seidel(A=A, b=b, x0=x2, tol=tol, max_iter=max_iterations) #x2

x0_vals_jacobi = [xi[0][0] for xi in tray_jacobi] #xo
x1_vals_jacobi = [xi[1][0] for xi in tray_jacobi]  #xo

x0_vals_seidel = [xi[0][0] for xi in tray_seidel] #xo
x1_vals_seidel = [xi[1][0] for xi in tray_seidel] #xo

x0_vals_jacobi1 = [xi[0][0] for xi in tray_jacobi1] #x2
x1_vals_jacobi1 = [xi[1][0] for xi in tray_jacobi1] #x2

x0_vals_seidel1 = [xi[0][0] for xi in tray_seidel1] #x2
x1_vals_seidel1 = [xi[1][0] for xi in tray_seidel1] #x2



# Grafica Gauss-Jacobi x0
plt.figure(figsize=(8, 6))
plt.plot(x0_vals_jacobi, x1_vals_jacobi, linestyle='--', marker='.', label='Iteraciones Gauss-Jacobi')
plt.scatter(tray_jacobi[0][0][0], tray_jacobi[0][1][0], color='blue', marker='s', s=150, label='Punto inicial', edgecolors='black', linewidths=2)
plt.scatter(x_jacobi[0][0], x_jacobi[1][0], color='red', marker='o', s=150, label='Solucin final Jacobi', edgecolors='black', linewidths=2)
plt.xlabel('x[0]')
plt.ylabel('x[1]')
plt.title('Mtodo de Gauss-Jacobi')
plt.grid(True)
plt.legend()
plt.show()

# Grafica Gauss-Seidel x0
plt.figure(figsize=(8, 6))
plt.plot(x0_vals_seidel, x1_vals_seidel, linestyle='--', marker='x', label='Iteraciones Gauss-Seidel')
plt.scatter(tray_seidel[0][0][0], tray_seidel[0][1][0], color='yellow', marker='s', s=150, label='Punto inicial', edgecolors='black', linewidths=2)
plt.scatter(x_seidel[0][0], x_seidel[1][0], color='black', marker='o', s=150, label='Solucin final Seidel', edgecolors='black', linewidths=2)
plt.xlabel('x[0]')
plt.ylabel('x[1]')
plt.title('Mtodo de Gauss-Seidel')
plt.grid(True)
plt.legend()
plt.show()

# Grafica Gauss-Jacobi x2
plt.figure(figsize=(8, 6))
plt.plot(x0_vals_jacobi1, x1_vals_jacobi1, linestyle='--', marker='.', label='Iteraciones Gauss-Jacobi')
plt.scatter(tray_jacobi1[0][0][0], tray_jacobi1[0][1][0], color='blue', marker='s', s=150, label='Punto inicial', edgecolors='black', linewidths=2)
plt.scatter(x_jacobi1[0][0], x_jacobi1[1][0], color='red', marker='o', s=150, label='Solucin final Jacobi', edgecolors='black', linewidths=2)
plt.xlabel('x[0]')
plt.ylabel('x[1]')
plt.title('Mtodo de Gauss-Jacobi')
plt.grid(True)
plt.legend()
plt.show()

# Grafica Gauss-Seidel x2
plt.figure(figsize=(8, 6))
plt.plot(x0_vals_seidel1, x1_vals_seidel1, linestyle='--', marker='x', label='Iteraciones Gauss-Seidel')
plt.scatter(tray_seidel1[0][0][0], tray_seidel1[0][1][0], color='yellow', marker='s', s=150, label='Punto inicial', edgecolors='black', linewidths=2)
plt.scatter(x_seidel1[0][0], x_seidel1[1][0], color='black', marker='o', s=150, label='Solucin final Seidel', edgecolors='black', linewidths=2)
plt.xlabel('x[0]')
plt.ylabel('x[1]')
plt.title('Mtodo de Gauss-Seidel')
plt.grid(True)
plt.legend()
plt.show()

#Grafifa convergencia 
A = np.array([[3, 2], [2, 10]], dtype=float)
b = np.array([[11], [16]], dtype=float)

x0 = np.array([[1], [1]], dtype=float)  
tol = 1e-6  
max_iter = 14  

x_jacobi, tray_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)
solution = np.array([x_jacobi[0, 0], x_jacobi[1, 0]])

tray_jacobi_points = np.array([t.flatten() for t in tray_jacobi[:13]])

plt.figure(figsize=(8, 6))

plt.plot(tray_jacobi_points[:, 0], tray_jacobi_points[:, 1], 'o-', label="Trayectoria Gauss-Jacobi", color="blue")

plt.scatter(tray_jacobi_points[0, 0], tray_jacobi_points[0, 1], color='red', label="$x_0 = (1, 1)$", zorder=5)
plt.scatter(solution[0], solution[1], color='green', label=f"Solucin estimada $x^* = ({solution[0]:.2f}, {solution[1]:.2f})$", zorder=5)

for point in tray_jacobi_points:
    plt.plot([point[0], solution[0]], [point[1], solution[1]], linestyle="dotted", color="gray", alpha=0.5)

plt.title("Trayectoria del Mtodo de Gauss-Jacobi (Iteraciones 0-12)")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.grid(alpha=0.5)
plt.legend()
plt.show()
```
